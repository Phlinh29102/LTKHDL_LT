### Với mỗi cột có kiểu dữ liệu dạng số (numerical), các giá trị được phân bố như thế nào?


Ở dữ liệu này, hiện có 6 cột có vẻ là thuộc nhóm **numeric**: "Population density", "Average population", "Sex ratio", "Population grow ratio", "15+ labor", "Year". Tuy nhiên, ta chỉ xếp 5 cột gồm "Population density", "Average population", "Sex ratio", "Population grow ratio", "15+ labor" vào nhóm **numeric**. Vì cột "Year" thật ra thuộc nhóm **categorical** do đây là dữ liệu dùng để phân biệt dữ liệu của một tỉnh thành ở nhiều năm khác nhau. 



```python
numerical_cols = df[['Population density', 'Average population', 'Sex ratio', 'Population grow ratio','15+ labor']].copy()
numerical_cols
```

#### Số lượng/tỉ lệ các giá trị thiếu?
#### Min? max? → Có gì bất thường không?
Với mỗi cột numeric ta sẽ tính tỉ lệ % giá trị thiếu (từ 0 đến 100), min, max. Và lưu kết quả vào dataframe `nume_col_profiles_df`.
Dataframe này gồm có:
- 3 dòng là "missing_ratio", "min", "max" lần lượt là tỉ lệ các giá trị thiếu mỗi cột, min và max.
- 5 cột là "Population density", "Average population", "Sex ratio", "Population grow ratio", "15+ labor".

```python
def missing_ratio(c):
    return  c.isna().sum()/len(df) * 100

def min(c):
    return c.min()

def max(c):
    return c.max()

nume_col_profiles_df = numerical_cols.agg([missing_ratio, min, max])
nume_col_profiles_df
```
